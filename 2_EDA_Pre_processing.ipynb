```python
# Import Libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load Your Data
# This assumes 'simulated_eon_grid_data.csv' is in the same folder
df = pd.read_csv('simulated_eon_grid_data.csv', index_col=0, parse_dates=True)

# Display Basic Info
print("First 5 rows:")
print(df.head())
print("\nData Info:")
df.info()
print("\nDescriptive Statistics:")
print(df.describe())

# Check for Missing Values
print("\nMissing Values per Column:")
print(df.isnull().sum())

# Plot a Single Day (to see daily solar impact clearly)
plt.figure(figsize=(12, 6))
df.loc['2024-07-15', ['Base_Demand_MW', 'High_Solar_MW', 'Net_Load_High_Solar_MW']].plot()
plt.title('Daily Load Profile with High Solar (July 15, 2024)')
plt.xlabel('Hour of Day')
plt.ylabel('Power (MW)')
plt.legend(['Base Demand', 'High Solar Generation', 'Net Load with High Solar'])
plt.grid(True)
plt.tight_layout()
plt.show()

# Plot 'Duck Curve' Effect (Average for a Summer Week)
summer_week_df = df.loc['2024-07-01':'2024-07-07']
hourly_avg = summer_week_df[['Base_Demand_MW', 'High_Solar_MW', 'Net_Load_High_Solar_MW']].groupby(summer_week_df.index.hour).mean()

plt.figure(figsize=(10, 6))
hourly_avg['Base_Demand_MW'].plot(label='Average Base Demand', color='blue')
hourly_avg['High_Solar_MW'].plot(label='Average High Solar Generation', color='gold')
hourly_avg['Net_Load_High_Solar_MW'].plot(label='Average Net Load (High Solar)', color='red', linestyle='--')
plt.title('Average Daily Load Profile & Duck Curve Effect (High Solar, Summer Week)')
plt.xlabel('Hour of Day')
plt.ylabel('Average Power (MW)')
plt.xticks(range(0, 24))
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

# Relationship between Temperature and Demand (Simple Scatter)
daily_df = df.resample('D').mean()
plt.figure(figsize=(10, 6))
sns.scatterplot(x=daily_df.index.dayofyear, y=daily_df['Base_Demand_MW'], hue=daily_df.index.month, palette='viridis')
plt.title('Daily Average Base Demand by Day of Year and Month')
plt.xlabel('Day of Year')
plt.ylabel('Average Base Demand (MW)')
plt.legend(title='Month')
plt.grid(True)
plt.show()
```
