```python
# Import Libraries
import pandas as pd
import numpy as np
from datetime import datetime, timedelta
import matplotlib.pyplot as plt

# Define Timeframe (2 years of hourly data)
start_date = datetime(2023, 1, 1)
end_date = datetime(2024, 12, 31, 23, 0, 0)
freq = 'H'
time_index = pd.date_range(start=start_date, end=end_date, freq=freq)
df = pd.DataFrame(index=time_index)

# Simulate Base Grid Demand (without solar)
df['hour'] = df.index.hour
df['day_of_year'] = df.index.dayofyear
df['day_of_week'] = df.index.dayofweek

base_load_mw = 1000

daily_factor = (np.sin(df['hour'] * 2 * np.pi / 24 - np.pi/2) * 0.2 +
                np.sin((df['hour'] - 18) * 2 * np.pi / 24) * 0.1)

seasonal_factor = np.sin(df['day_of_year'] * 2 * np.pi / 365 * 2) * 0.15

weekly_factor = np.where((df['day_of_week'] == 5) | (df['day_of_week'] == 6), 0.85, 1.0)

noise = np.random.normal(0, 30, len(df))

df['Base_Demand_MW'] = base_load_mw * (1 + daily_factor + seasonal_factor) * weekly_factor + noise
df['Base_Demand_MW'] = df['Base_Demand_MW'].clip(lower=0)

# Simulate Residential Solar Generation
avg_panel_size_kw = 5

solar_daily_pattern = np.sin((df['hour'] - 6) * np.pi / 12).clip(lower=0)

solar_seasonal_pattern = (np.sin(df['day_of_year'] * 2 * np.pi / 365 - np.pi/2) + 1) / 2

df['Solar_Output_kW_per_home'] = solar_daily_pattern * solar_seasonal_pattern * avg_panel_size_kw
df['Solar_Output_kW_per_home'] = df['Solar_Output_kW_per_home'].clip(lower=0)

# Calculate Total Solar Generation for Different Penetration Levels
num_total_homes = 10000

solar_penetration_levels = {
    'No_Solar_MW': 0.0,
    'Low_Solar_MW': 0.10,
    'Medium_Solar_MW': 0.30,
    'High_Solar_MW': 0.60
}

for level_name, penetration_rate in solar_penetration_levels.items():
    num_solar_homes_at_level = int(num_total_homes * penetration_rate)
    df[level_name] = df['Solar_Output_kW_per_home'] * num_solar_homes_at_level / 1000

# Calculate Net Load for each scenario (Demand - Solar Generation)
df['Net_Load_No_Solar_MW'] = df['Base_Demand_MW'] - df['No_Solar_MW']
df['Net_Load_Low_Solar_MW'] = df['Base_Demand_MW'] - df['Low_Solar_MW']
df['Net_Load_Medium_Solar_MW'] = df['Base_Demand_MW'] - df['Medium_Solar_MW']
df['Net_Load_High_Solar_MW'] = df['Base_Demand_MW'] - df['High_Solar_MW']

# Visualize a Sample Week (to check your data)
plt.figure(figsize=(16, 8))
plt.plot(df.loc['2024-07-01':'2024-07-07', 'Base_Demand_MW'], label='Base Demand (No Solar)', color='blue')
plt.plot(df.loc['2024-07-01':'2024-07-07', 'Net_Load_Low_Solar_MW'], label='Net Load (Low Solar)', color='orange')
plt.plot(df.loc['2024-07-01':'2024-07-07', 'Net_Load_Medium_Solar_MW'], label='Net Load (Medium Solar)', color='green')
plt.plot(df.loc['2024-07-01':'2024-07-07', 'Net_Load_High_Solar_MW'], label='Net Load (High Solar)', color='red')
plt.title('Simulated Grid Load Profiles with Varying Solar Penetration (Week in July)')
plt.xlabel('Date and Time')
plt.ylabel('Power (MW)')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

# Save Your Simulated Data
final_df = df[['Base_Demand_MW', 'No_Solar_MW', 'Low_Solar_MW', 'Medium_Solar_MW', 'High_Solar_MW',
               'Net_Load_No_Solar_MW', 'Net_Load_Low_Solar_MW', 'Net_Load_Medium_Solar_MW', 'Net_Load_High_Solar_MW']]
final_df.to_csv('simulated_eon_grid_data.csv', index=True)
print("Simulated data saved to simulated_eon_grid_data.csv")
```
